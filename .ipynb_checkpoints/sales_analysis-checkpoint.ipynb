```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset with correct encoding and clean column names
df = pd.read_csv("sales_data.csv", encoding='cp1252')
df.columns = df.columns.str.strip()

# Convert order date to datetime format
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
```

```python
product_sales = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 5))
product_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Product Line')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Find top 5 countries by total sales
top_countries = df.groupby('COUNTRY')['SALES'].sum().sort_values(ascending=False).head(5)

# Create a pivot table for heatmap
pivot_geo = df[df['COUNTRY'].isin(top_countries.index)].pivot_table(
    index='PRODUCTLINE',
    columns='COUNTRY',
    values='SALES',
    aggfunc='sum'
)

# Visualize heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(pivot_geo, annot=True, fmt=".0f", cmap="Blues")
plt.title('Sales by Product Line and Country (Top 5)')
plt.tight_layout()
plt.show()
```

```python
# Extract year-month from order date
df['Month'] = df['ORDERDATE'].dt.to_period('M')

# Aggregate sales by month
monthly_sales = df.groupby('Month')['SALES'].sum()

# Plot monthly sales trend
plt.figure(figsize=(10, 4))
monthly_sales.plot(marker='o')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Calculate unit price for each transaction
df['UnitPrice'] = df['SALES'] / df['QUANTITYORDERED']

# Average unit price per product line
avg_price = df.groupby('PRODUCTLINE')['UnitPrice'].mean().sort_values(ascending=False)

# Plot average unit prices
plt.figure(figsize=(8, 5))
avg_price.plot(kind='bar', color='orange')
plt.title('Average Unit Price by Product Line')
plt.ylabel('Price ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: quantity ordered vs. sales
plt.figure(figsize=(6, 4))
sns.scatterplot(data=df, x='QUANTITYORDERED', y='SALES', alpha=0.5)
plt.title('Quantity Ordered vs Sales')
plt.xlabel('Quantity Ordered')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
